```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python

year = 2018# float('nan')
week = 1#float('nan')
df = pd.DataFrame([], columns= ["date", "homeTeam", "homeTeamPTS", "awayTeam",'awayTeamPTS','Over/Under'])
temp_arr = []
for week in range(1,18):
    url = 'https://www.pro-football-reference.com/years/'+str(year)+'/week_'+str(week)+'.htm'
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    soup1 = soup.find_all('div', class_='game_summary expanded nohover')
    for el in soup1:
        soup2 = el.find_all('table', class_='teams')
        for el2 in soup2:
            soup3 = el2.find_all('td')
            date = str(re.findall('>[A-Z].*<', str(soup3[0]))[0])[1:-1]
            team1 = str(re.findall('>[A-Z][a-z]* [A-Z][a-z]*',str(soup3[1]))[0])[1:]
            team1_points = str(re.findall('>\d+',str(soup3[2]))[0])[1:]
            ou_url = 'https://www.pro-football-reference.com/boxscores/'+str(re.findall('\d{9}.+.htm',str(soup3[3]))[0])
            r2 = requests.get(ou_url)
            soup4 = BeautifulSoup(r2.content, 'html.parser')
            soup5 = soup4.find('div',class_="content_grid")
            soup6 = str(soup5.find('div'))
            ou = re.findall('\d.+ ', str(re.findall('Over/Under.* ',soup6)))[0][:-1]
            temp_arr.append(ou_url)
            team2 = str(re.findall('>[A-Z][a-z]* [A-Z][a-z]*',str(soup3[4]))[0])[1:]
            team2_points = str(re.findall('>\d+',str(soup3[5]))[0])[1:]
            temp_series = pd.Series([date,team2,team2_points, team1,team1_points, ou], df.columns)
            df = df.append(temp_series, ignore_index=True)
display(df)


```
